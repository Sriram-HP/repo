```python
df.groupby('Sub-Category')['Profit'].describe()
```

```python
#repeat this for all the columns and give insights
```

```python
#top 3 sub-categories on the basis of means
df.groupby('Sub-Category')['Profit'].mean().sort_values(ascending=False).head(3)
```

```python
#bottom 3 sub-categories on the basis of count
df.groupby('Sub-Category')['Profit'].count().sort_values(ascending=True).head(3)
```

```python
df.groupby('Sub-Category')['Profit'].count().sort_values(ascending=False).tail(3)
```

```python
import seaborn as cat
```

```python
cat.catplot(
    data=df, kind="bar",
    y="Sub-Category", x="Profit")
```

```python
import matplotlib.pyplot as plt
f, ax = plt.subplots(figsize=(10, 10))

cat.scatterplot(x="Sales", y="Profit", data=df, hue='Sub-Category')
#can i color it by category
```
